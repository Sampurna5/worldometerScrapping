```python
from bs4 import BeautifulSoup
import requests

list_data = []
final_data = []

url = BeautifulSoup('https://www.worldometers.info/coronavirus/', 'html.parser')
response = requests.get(url)

data = response.text

xmlData = BeautifulSoup(data, 'lxml')

tags = xmlData.table

tag = tags.find_all('tr')


# deletes table rows with unwanted data mentioned in delIndexes

delIndexes = [1, 2, 3, 4, 5, 6, 7, 8, 230, 231, 232, 233, 234, 235, 236]
for index in sorted(delIndexes, reverse = True):
    del tag[index]

for i in range(len(tag)):
    list_data.append(tag[i].get_text())

for i in list_data:
    final_data.append(i.split('\n')[1:-3])
    

# inserts China row in it's actual index i.e. 83 instead at the end of the table and corrects every
# indexes after it

final_data.insert(83, final_data[221])
del final_data[222]
for data in final_data[83:-1]:
    if data[0] == '221':
        data[0] = '83'
    else:
        sn = data[0]
        sn = int(sn)
        sn += 1
        data[0] = str(sn)


# joins table header at index [13] i.e. 'Tests' and [14] i.e. '/1M pop' in index [13] and deletes
# extra '/1Mpop' at index [14] and deletes extra empty cell at index [14]

final_data[0][13] = final_data[0][13] + final_data[0][14]
del final_data[0][14:16]
final_data[-1] = final_data[-1][:15]
```

```python
import pandas as pd

dataFrame = pd.DataFrame(final_data)
dataFrame.to_csv('COVID-Updates.csv', mode = 'w', header = False, index = None)

dataFrame = pd.read_csv('COVID-Updates.csv', index_col = '#')
dataFrame.fillna('-', inplace = True)
dataFrame
```
